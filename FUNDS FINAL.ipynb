```python
def exchange(transfer,amount,recieve):
    for i in range(len(data)):
        for j in range(2):
            if(data[i][j]==transfer):
                data[i][j+1]=int(data[i][j+1])-amount
                x=data[i][j+1]
            if(data[i][j]==recieve):
                data[i][j+1]=int(data[i][j+1])+amount
                y=data[i][j+1]
    with open('track','a',encoding='utf-8') as x:
            x.write('%s transfered %s to %s \n'%(transfer,amount,recieve))
    return x,y
def new_registration():
    print('Enter your name\nEnter the amount to be stored in account\nNote: Enter details giving spaces\n')
    B=list(map(str,input().split()))
    if B[0] in (item for sublist in data for item in sublist):
        print('You are already registered\n')
        return
    data.append(B)
    return data
def transfer_amount(A):
    res1=A[0] in (item for sublist in data for item in sublist) 
    res2=A[2] in (item for sublist in data for item in sublist)
    if res1==True:
        print('You have successfully logged in')
    if res1 and res2: 
        t=exchange(A[0],int(A[1]),A[2])
        print('Transaction succesful')
    if res1 and res2==False:
        print('Invalid transfer')
    if res1==False:
        print(A[0] +' kindly register yourself')
        new_registration()
    if res2==False:
        print(A[2] +' is not a registered member of our bank')
```
